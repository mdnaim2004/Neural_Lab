## Tuple

A tuple is an ordered immutable collection in python. tuples are writen with parenthease().

tuple characterastrics:

- Orderd --> element have a define orderd
- immutable --> cannot change element after creation.
- cannot duplicate
- can contain diffrent data type


```python
# tuple define

tpl = (1, 2, 3)
print(type (tpl))

a_list = [10, 20, 30, 40, 50, 60, 70, 80, 90]
print(type(a_list))

# type cast

a_tuple = tuple(a_list)
print(type(a_tuple))
print(a_tuple)
```
