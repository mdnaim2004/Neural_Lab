## Attribute and Methode

```python
class car:
    def __init__(self, brand, color):
        self.brand = brand
        self.color = color
    

my_car = car("BMW", "Black")

print(my_car.brand)
print(my_car.color)
```

#### Methode

```python
class car:
    def __init__(self, brand, color):
        self.brand = brand
        self.color = color

    def drive(self):
        print(f"The {self.brand} is driving")

car1 = car("BMW", "Black")
car1.drive()
```

```python
class Student:
    def __init__(self, name, id, gender, sub, dept):
        self.name = name
        self.id = id
        self.gender = gender
        self.sub = sub
        self.dept = dept

    def study(self):
        print(f"{self.name} is studying {self.sub}")

    def info(self):
        print(f"Name : {self.name} \n ID : {self.id} \n Gender : {self.gender} \n SUB : {self.sub} \n Dept : {self.dept}")
```

```python
student1 = Student("Naim", 68, "Male", "CSE", "S&E")
```

```python
student1.study()
```

```python
student1.info()
```
