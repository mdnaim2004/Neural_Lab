## Introduction to OOPs (Object Orentade Programing system)

#### OOPs is a programing paradigm based based on:
- Classes -- blueprint
- Objects -- instances
- Attributes -- variable inside a class
- Methodes -- function inside a class

#### Main OOP principles:
- **Encaptulation** : Keeping data and methos together in a class and restracting direc access to internal data.
- **Inheritance** : One class drives from another, gainig its properties and behaviour(code reuse)
- **Polymorphism** : Same methode name behaving differently depending on the object calling it.
- **Abstrution** : Showing only essential feature while hiding unnecessary implemention details.
