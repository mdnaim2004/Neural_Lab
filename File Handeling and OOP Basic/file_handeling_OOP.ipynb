Openning file

```python
file = open("Example.txt", "r") # open a file for reading
```

### Some operation
- "r" ==> read --> file must exist
- "w" ==> write --> create new file; overwrites existing
- "a" ==> append --> add the data to end to end file
- "r+" ==> read + write --> file must exist
- "w+" ==> write + read --> overwrite files
- "a+" ==> append + read --> create if not exist.
