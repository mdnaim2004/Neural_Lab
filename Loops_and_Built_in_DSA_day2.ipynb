### While Loop

## break

- break statement means program stoped of spwcific conditions,


```python
for i in range(10):
    print(i)
    if(i == 5):
        break

print("i am now out of the loop")
```

```python
for i in range(10):
    print(i)
    break
```

```python
for i in range(1, 20, 1):
    print(i)
    if i == 5:
        break
```

```python
for i in range(2, 20, 2):
    print(i)
    if i == 5:
        break
```

## Continue

```python
for i in range(10):
    if ( i % 2 == 0):
        continue
    print(i)
```

```python
for i in range(10):
    if ( i % 2 == 0):
        print(i)
```
