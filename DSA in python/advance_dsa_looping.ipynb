### List

```python
fruits = ['apple', 'banana', 'mango']

# simple iterations....

for fruit in fruits:
    print(fruit)
```

####  With the indexicing print of the list...

```python
for i in range(len(fruits)):
    print(f"Index {i} : {fruits[i]}")
```
