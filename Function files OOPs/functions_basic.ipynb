## Defining Function


```python
def greet():
    print("hello World!")
```

```python
greet()
```

```python
def greet(name):
    print(f"hello hi {name}")
```

```python
greet("naim")
```

```python
def func(name):
    print(f"hello {name}")
```

```python
func("naim islam")
```

```python
def fun(name, age):
    print(f" hello {name}, your age is = {age}")
```

```python
fun("naim", 22)
```
